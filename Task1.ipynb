```python
cleaned_data.isnull()
```

```python
# creating a plot where x=Ram types & y=their number
x=cleaned_data['RAM'].unique().tolist()
print(x)
y=cleaned_data['RAM'].value_counts(sort=False).tolist()
print(y)
```

```python
fig,((ax1),(ax2))=plt.subplots(nrows=1,
                                     ncols=2,
                                     figsize=(10,10))
ax1.scatter(x,y);
ax2.bar(x,y);
```

```python
plt.pie(y, labels = x, autopct = '%1.1f%%',
 startangle = 90, 
 wedgeprops = {"edgecolor" : "white",
 'linewidth': 1,
 'antialiased': True})
plt.axis('equal')
 
# Display the graph onto the screen
plt.show()
```
